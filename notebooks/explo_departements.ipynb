### Éssaie de calcul de la consommation totale en fonction de la température du mois, de la température seuil, de la thermosensibilité (MWh/DJU)

```python
temperature_moyenne_mois = 5    # °C
temperature_seuil = 16           # °C (pour la Nièvre)
thermosensibilite = 127          # MWh/DJU (exemple trouvé pour Nièvre 2017)
nb_jours_mois = 31               # janvier


# consommation non thermosensible journalière moyenne
conso_non_thermosensible_jour = 379_114 / 365  # MWh/jour (exemple Nièvre 2017)

# === Calcul des DJU mensuels ===

dju_moyen_jour = max(0, temperature_seuil - temperature_moyenne_mois)
dju_total_mois = dju_moyen_jour * nb_jours_mois

# === Calcul consommation thermosensible ===

conso_thermosensible_mois = thermosensibilite * dju_total_mois

# === Consommation non thermosensible sur le mois ===

conso_non_thermosensible_mois = conso_non_thermosensible_jour * nb_jours_mois

# === Consommation totale estimée ===

conso_totale_estimee_mois = conso_thermosensible_mois + conso_non_thermosensible_mois

print(f"==> Consommation totale estimée : {conso_totale_estimee_mois:.0f} MWh pour le mois")
```

### Évolution de la thermosensibilité dans le temps

```python
df_reg = df_selection_clean[["Code Département", "Année", "Thermosensibilité totale (kWh/DJU)"]].sort_values(['Code Département', 'Année'])

df_reg
```

### Calcul de la regression linéaire de la thermosensibilité au cours des années

```python
from scipy.stats import linregress

# Liste pour stocker les résultats
results = []

# Boucle sur chaque département
for dept, group in df_reg.groupby('Code Département'):
    if len(group) >= 2:  # au moins 2 points pour une régression
        slope, intercept, r_value, p_value, std_err = linregress(group['Année'], group['Thermosensibilité totale (kWh/DJU)'])
        results.append({
            'Code Département': dept,
            'a (slope)': slope,
            'b (intercept)': intercept,
            'r²': r_value**2,
            'n': len(group)
        })

# Mettre sous forme de DataFrame
df_models = pd.DataFrame(results)

# Trier pour voir les départements avec la plus forte pente (positive ou négative)
df_models_sorted = df_models.sort_values('a (slope)', ascending=False)


print(df_models_sorted.head(2))  # les plus fortes hausses
print(df_models_sorted.tail(2))  # les plus fortes baisses

# export de df_regression_thermosensibilité en csv
df_models_sorted.to_csv("regression-thermosensibilite-departements.csv", sep=';', index=False)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math

# Ton DataFrame filtré déjà prêt
df_plot = df_selection_clean.dropna(subset=['Thermosensibilité totale (kWh/DJU)'])

# Liste complète des départements
departements = df_plot['Code Département'].unique()
n_departements = len(departements)

# Trouver min et max globaux pour fixer l'échelle
ymin = df_plot['Thermosensibilité totale (kWh/DJU)'].min()
ymax = df_plot['Thermosensibilité totale (kWh/DJU)'].max()

# Ajuster un peu pour marges
padding = (ymax - ymin) * 0.05
ymin -= padding
ymax += padding

# Organiser la grille
ncols = 5
nrows = math.ceil(n_departements / ncols)

plt.figure(figsize=(4 * ncols, 4 * nrows))

for idx, dept in enumerate(departements, 1):
    plt.subplot(nrows, ncols, idx)
    
    dept_data = df_plot[df_plot['Code Département'] == dept].sort_values('Année')
    
    plt.plot(
        dept_data['Année'],
        dept_data['Thermosensibilité totale (kWh/DJU)'],
        color='blue',
        linestyle='-',
        marker='o',
        label="Données"
    )
    
    row = df_models_sorted[df_models_sorted['Code Département'] == dept].iloc[0]
    a = row['a (slope)']
    b = row['b (intercept)']
    r2 = row['r²']
    
    years = np.arange(2017, 2026)
    predicted = a * years + b
    
    plt.plot(
        years, 
        predicted,
        color='orange',
        linestyle='--',
        label=f"Régression (r²={r2:.2f})"
    )
    
    # Fixer la même échelle Y pour tous
    plt.ylim(ymin, ymax)
    
    plt.title(f"Département {dept}")
    plt.xlabel("Année")
    plt.ylabel("Thermosensibilité totale (kWh/DJU)")
    plt.grid(True)
    
    if idx <= 3:
        plt.legend()

plt.tight_layout()
plt.show()
```

### fonction d'estimation de la consommation totale en fonction de la température du mois, de la température seuil, de la thermosensibilité (MWh/DJU)

```python
def estimer_thermosensibilite(departement_code, annee):
    row = df_models_sorted[df_models_sorted['Code Département'] == departement_code]
    if row.empty:
        return None, f"Aucune régression trouvée pour le département {departement_code}."
    
    a = row['a (slope)'].values[0]
    b = row['b (intercept)'].values[0]
    r2 = row['r²'].values[0]
    
    thermosensibilite = a * annee + b  # en kWh/DJU
    return thermosensibilite, f"Thermosensibilité estimée = {thermosensibilite:.0f} kWh/DJU (r²={r2:.2f})"

def estimer_conso_thermo(thermosensibilite, departement_code, temp_moyenne_mois, nb_jours=31, seuil=16):
    # Calculer DJU pour le mois
    dju_mois = max(0, seuil - temp_moyenne_mois) * nb_jours
    
    # Estimer conso thermosensible en MWh
    conso_thermo = (thermosensibilite * dju_mois) / 1000  # conversion en MWh
    
    return (f"Pour {nb_jours} jours à {temp_moyenne_mois}°C,"
            f"\nDJU du mois = {dju_mois:.0f},"
            f"\n=> Consommation usages thermosensibles estimée ≈ {conso_thermo:,.0f} MWh")

# Exemple d'utilisation
departement_code = '58'  # Nièvre
annee = 2025
thermosensibilite, message = estimer_thermosensibilite(departement_code, annee)
print(message)

temp_moyenne_mois = 5 # °C
conso_thermo = estimer_conso_thermo(thermosensibilite, departement_code, temp_moyenne_mois)
print(conso_thermo)

def estimer_conso_complete_mensuelle(departement_code, annee, temp_moyenne_mois, nb_jours=31, seuil=16):
    # Estimer la thermosensibilité projetée
    thermo, info = estimer_thermosensibilite(departement_code, annee)
    if thermo is None:
        return info
    
    # Calculer DJU pour le mois
    dju_mois = max(0, seuil - temp_moyenne_mois) * nb_jours
    
    # Conso usages thermosensibles du mois
    conso_thermo_mois = (thermo * dju_mois) / 1000  # MWh
    
    # Conso usages non thermosensibles du mois
    data_non_thermo = df_selection_clean[df_selection_clean['Code Département'] == departement_code]
    conso_non_thermo_annee_moy = data_non_thermo['Conso totale  usages non thermosensibles (MWh)'].mean()
    conso_non_thermo_mois = conso_non_thermo_annee_moy / 12  # hypothèse homogène sur l'année
    
    # Conso totale du mois
    conso_totale_mois = conso_thermo_mois + conso_non_thermo_mois
    
    return (f"Pour le département {departement_code} en {annee},"
            f"\npour un mois de {nb_jours} jours à {temp_moyenne_mois}°C :"
            f"\n- DJU du mois = {dju_mois:.0f}"
            f"\n- Conso usages thermosensibles ≈ {conso_thermo_mois:,.0f} MWh"
            f"\n- Conso usages non thermosensibles ≈ {conso_non_thermo_mois:,.0f} MWh"
            f"\n=> Consommation totale estimée du mois ≈ {conso_totale_mois:,.0f} MWh")

# Exemple d'utilisation
departement_code = '58'  # Nièvre
annee = 2025
temp_moyenne_mois = 5

resultat = estimer_conso_complete_mensuelle(departement_code, annee, temp_moyenne_mois)
print(resultat)


```

### Évolution de la consommation totale au cours des années

```python
import numpy as np
import matplotlib.pyplot as plt
import math
from scipy.stats import linregress

# Préparer le dataframe
df_non_thermo = df_selection_clean.dropna(subset=['Conso totale  usages non thermosensibles (MWh)'])

# Liste complète des départements
departements = df_non_thermo['Code Département'].unique()
n_departements = len(departements)

# Trouver min et max globaux pour fixer l'échelle Y
ymin = df_non_thermo['Conso totale  usages non thermosensibles (MWh)'].min()
ymax = df_non_thermo['Conso totale  usages non thermosensibles (MWh)'].max()
padding = (ymax - ymin) * 0.05
ymin -= padding
ymax += padding

# Organiser la grille
ncols = 5
nrows = math.ceil(n_departements / ncols)

plt.figure(figsize=(4 * ncols, 4 * nrows))

for idx, dept in enumerate(departements, 1):
    plt.subplot(nrows, ncols, idx)
    
    dept_data = df_non_thermo[df_non_thermo['Code Département'] == dept].sort_values('Année')
    
    # Ligne continue pour les données historiques
    plt.plot(
        dept_data['Année'],
        dept_data['Conso totale  usages non thermosensibles (MWh)'],
        color='blue',
        linestyle='-',
        marker='o',
        label="Données"
    )
    
    # Calculer régression linéaire sur cette série
    slope, intercept, r_value, p_value, std_err = linregress(
        dept_data['Année'], dept_data['Conso totale  usages non thermosensibles (MWh)'])
    
    years = np.arange(dept_data['Année'].min(), 2026)
    predicted = slope * years + intercept
    
    # Ligne pointillée pour la régression
    plt.plot(
        years, 
        predicted,
        color='orange',
        linestyle='--',
        label=f"Régression (r²={r_value**2:.2f})"
    )
    
    # Fixer la même échelle Y pour tous
    plt.ylim(ymin, ymax)
    
    plt.title(f"Département {dept}")
    plt.xlabel("Année")
    plt.ylabel("Conso non thermosensible (MWh)")
    plt.grid(True)
    
    if idx <= 3:
        plt.legend()

plt.tight_layout()
plt.show()
```
